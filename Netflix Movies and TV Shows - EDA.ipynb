## data cleaning

```python
# Step 1: Drop the 'show_id' column
df.drop(columns=['show_id'], inplace=True)
df.drop(columns=['description'], inplace=True)
```

```python
df.duplicated().sum()
df.drop_duplicates()
df.head()
```

```python
df.isna().sum()
```

```python

# Step 2: Handle missing values
df['director'].fillna('Unknown', inplace=True)
df['casts'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Not Rated', inplace=True)
df['duration'].fillna('Unknown', inplace=True)
df.dropna(subset=['date_added'], inplace=True)  # drop rows with missing date_added
```

```python
df
```

```python
# Step 3: Convert date_added to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df.rename(columns={'listed_in':'categories'},inplace=True)
df.head()
```

```python
df['duration_min'] = df['duration'].apply(
    lambda x: int(str(x).split()[0]) if isinstance(x, str) and 'min' in x else 0)

df['duration_season'] = df['duration'].apply(
    lambda x: int(str(x).split()[0]) if isinstance(x, str) and 'Seasons' in x else 0
)
```

## visualization

```python
# Plot the distribution of top 10 countries 

plt.figure(figsize=(12, 6))
sns.countplot(x='country', data=df, order=df['country'].value_counts().head(10).index)
plt.xticks(rotation=90)
plt.title('Distribution of Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the trend of movies and shows over time

plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='month', hue='type', data=df)
plt.title('Trend of Movies and Shows Over Time')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

```python
# Plot the average duration by category (Top 15)
plt.figure(figsize=(12, 6))
df.groupby("categories")["duration_min"].mean().sort_values(ascending=False).head(15).plot(kind='bar')
plt.title('Average Duration by Category (Top 30)')
plt.xlabel('Category')
plt.ylabel('Average Duration (minutes)')
plt.xticks(rotation=90)
plt.show()
```

```python
# 1. Content Type Distribution
plt.subplot(2,1,2)
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Content Type Distribution')
```

```python
# 2. Content Added Over Time
plt.subplot(2, 1, 2)
df['year'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title('Content Added by Year')
plt.xlabel('Year')
plt.ylabel('Count')
```

```python
# 3. Release Year Distribution
plt.subplot(2, 1, 2)
sns.histplot(data=df, x='release_year', bins=30, kde=True)
plt.title('Release Year Distribution')
plt.xlabel('Release Year')
```

```python
# 4. Top Production Countries
plt.subplot(2, 1, 2)
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Production Countries')
plt.xlabel('Count')
```

```python
# 6. Duration Analysis
plt.subplot(2, 1, 2)
movies = df[df['type'] == 'Movie']
sns.histplot(data=movies, x='duration_min', bins=30, kde=True)
plt.title('Movie Duration Distribution (minutes)')
plt.xlabel('duration (min)')
plt.tight_layout()
plt.show()
```

```python
## TV Shows Analysis
tv_shows = df[df['type'] == 'TV Show']

plt.figure(figsize=(12, 6))
sns.countplot(data=tv_shows, x='duration_min', palette='Set2')
plt.title('TV Show Seasons Count')
plt.xlabel('Number of Seasons')
plt.ylabel('Count')
plt.show()
```
